```python

import numpy as np
import cv2 
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

```python
img = cv2.imread("tictac.jpg")
#img = cv2.resize(img, dsize=(512, 512), interpolation=cv2.INTER_CUBIC)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
def show_diff(kernel):
  convolved = convolve2d(img, kernel)
  fig = plt.figure(figsize=(10,10))
  plt.subplot(121)
  plt.title("original")
  plt.axis("off")
  plt.imshow(img, cmap = "gray")

  plt.subplot(122)
  plt.title("convolved")
  plt.axis("off")
  plt.imshow(convolved, cmap = "gray")
  return convolved
```

```python
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])/128

kernel2 = np.ones((8,8),np.float32)/64
output = show_diff(kernel2)
```

```python
kernel3 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], np.float32)

dx = show_diff(kernel3)
```

```python
kernel4 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]], np.float32)

dy = show_diff(kernel4)
```

```python
tetris = cv2.imread("tetris.jpg")
#Print out the type of image data and its dimensions ( height , width , and color)
print("This image is: ", type(tetris), "with dimensions:", tetris.shape)

tetris_copy =np.copy(tetris)

tetris_copy = cv2.cvtColor( tetris_copy, cv2.COLOR_BGR2RGB)

plt.imshow(tetris_copy,cmap="gray")
```

```python
edges = cv2.Canny(tetris_copy,30,150)

fig = plt.figure(figsize=(10,10))
plt.imshow(edges, cmap="gray")
```

```python
tetris_copy = cv2.cvtColor( tetris_copy, cv2.COLOR_BGR2GRAY)
#La funcion de threshold vuelve los pixeles con valores mas altos a 225 en este caso como relevantes
threshold = cv2.threshold(tetris_copy, 225, 255, cv2.THRESH_BINARY_INV)[1]
plt.imshow(threshold, cmap = "gray")
```
