```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("3.jpg")

img = cv2.resize(img, dsize=(512, 512), interpolation=cv2.INTER_CUBIC)
plt.imshow(img)
```

```python
print("This image is: ", type(img))
print("with dimensions: ", img.shape)
```

```python
img_copy = np.copy(img)
```

```python
img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
print(img_copy.shape)
plt.imshow(img_copy, cmap="gray")
```

```python
examplePixels = [[240,190],[70,270],[380,300],[240,310],[400,40]]

examples = []

```

```python
for pixel in examplePixels:
  example = img_copy[pixel[0]-16:pixel[0]+16,pixel[1]-16:pixel[1]+16]
  examples.append(example)

fig = plt.figure(0, (6,4))

for i, example in enumerate(examples):
  ax = fig.add_subplot(2,3,i+1)
  ax.imshow(example, cmap= "gray", interpolation = "none",vmin = 0, vmax= 200)
  ax.set_xticks([])
  ax.set_yticks([])

```
